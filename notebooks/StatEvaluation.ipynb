```python
import numpy as np
import scipy.stats as stats
from statsmodels.stats.proportion import proportions_ztest


# Control Group Data
control_users = 5076
control_conversions = 503
control_sla_breaches = 27

# Variant Group Data
variant_users = 4924
variant_conversions = 627
variant_sla_breaches = 83

print("--- A/B TEST STATISTICAL REPORT ---")

# --- 2. Sample Ratio Mismatch (SRM) Check ---
observed = [control_users, variant_users]
expected = [sum(observed)/2, sum(observed)/2]
_, srm_p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print(f"\n1. SRM Check p-value: {srm_p_value:.4f}")
if srm_p_value < 0.05:
    print("Nah")
else:
    print("Yay")

# --- 3. Primary Metric: Conversion Rate Z-Test ---
count = np.array([control_conversions, variant_conversions])
nobs = np.array([control_users, variant_users])
z_stat, p_val = proportions_ztest(count, nobs, alternative='two-sided')

control_cr = control_conversions / control_users
variant_cr = variant_conversions / variant_users
relative_lift = (variant_cr - control_cr) / control_cr

print(f"\n2. Conversion Rate Lift:")
print(f"   Control CR: {control_cr:.2%}")
print(f"   Variant CR: {variant_cr:.2%}")
print(f"   Relative Lift: {relative_lift:.2%}")
print(f"   P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Amazing")
else:
    print("Nah")

# --- 4.SLA Breach Rate ---
control_sla_rate = control_sla_breaches / control_conversions
variant_sla_rate = variant_sla_breaches / variant_conversions

print(f"\n3. Guardrail Check (SLA Breaches):")
print(f"   Control SLA Breach Rate: {control_sla_rate:.2%}")
print(f"   Variant SLA Breach Rate: {variant_sla_rate:.2%}")
```
